```python
import pyromat as pm
import numpy as np
import matplotlib.pyplot as plt

pm.config["unit_pressure"] = "kPa"
pm.config["def_p"] = 100

mp_water = pm.get("mp.H2O")

p = np.arange(1.,22063.)
```

```python
T = mp_water.Ts(p=p)
s = mp_water.ss(p=p)
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(15,10))
#plt.xscale("log")
#plt.yscale("log")
plt.title('Saturation P-v Diagram for Water')
plt.rc('font', **font)
plt.plot(s[0],T, 'b--')
plt.plot(s[1],T,'r--')
plt.ylabel('Temperature (K)')
plt.xlabel('Enthropy (kJ)')
```
