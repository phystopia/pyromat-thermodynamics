## P-v Diagram

```python
p = np.arange(15,3390)
v = Nitrogen.vs(p=p)
vf= v[0]
vg= v[1]
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(30,10))
plt.xscale("log")
plt.yscale("log")
plt.title('P-v Diagram for Nitrogen')
plt.rc('font', **font)
#saturation diag
plt.plot(vf,p, 'b--')
plt.plot(vg,p,'r--')
#process
plt.plot([v1,v2],[P1,P2],'g',linewidth=2)
plt.plot([v2,v3],[P2,P3],'g',linewidth=2)

plt.ylabel('Pressure (kPa)')
plt.xlabel('specific volume (m³/kg)')
```

## Saturation T-$\upsilon$ Diagram for Nitrogen

### Defining saturation temperatures.

```python
T = np.arange(70,126.26)
```

```python
v = Nitrogen.vs(T=T)
vf= v[0]
vg= v[1]
```

### Creating the log-log graph.

```python
font = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(30,10))
plt.xscale("log")
plt.yscale("log")
plt.title('Saturation T-v Diagram for Nitrogen')
plt.rc('font', **font)
plt.plot(vf,T, 'b--')
plt.plot(vg,T,'r--')

plt.plot([v1,v2],[t1,t2],'g',linewidth=2)
plt.plot([v2,v3],[t2,t3],'g',linewidth=2)

plt.ylabel('Temperature (K)')
plt.xlabel('specific volume (m³/kg)')
```

### Creating the "normal" graph.

```python
ont = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(30,10))
plt.title('Saturation T-v Diagram for Nitrogen')
plt.rc('font', **font)
#plt.xlim([0,1])
plt.plot(vf,T, 'b--')
plt.plot(vg,T,'r--')
plt.ylabel('Temperature (K)')
plt.xlabel('specific volume (m³/kg)')
```

## Saturation T-s Diagram for Nitrogen

### New range of pressures, in order to not have errors calculating derivatives.

```python
p = np.arange(15,3390)
```

### Defining saturation temperatures and entropies.

```python
T = Nitrogen.Ts(p=p)
s = Nitrogen.ss(p=p)
```

### Creating the graph.

```python
font = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(15,10))
plt.title('Saturation T-s Diagram for Nitrogen')
plt.rc('font', **font)
plt.plot(s[0],T, 'b--')
plt.plot(s[1],T,'r--')
plt.plot([s1,s2],[t1,t2],'g',linewidth=2)
plt.plot([s2,s3],[t2,t3],'g',linewidth=2)
plt.ylabel('Temperature (K)')
plt.xlabel('specific enthropy (kJ/kg)')
```
