# Example 2.1 Evaluating Moving-Boundary Work.

A gas in a piston–cylinder assembly undergoes an expansion process for which the relationship between pressure and volume is given by
<br>

$$ P \cdot V^n = \mathrm{constant} $$
<br>
The initial pressure is $3 \, \mathrm{bar}$, the initial volume is $0.1 \, \mathrm{m^3}$, and the final volume is $0.2 \, \mathrm{m^3}$.
<br>
Determine the work for the process if:
<br>
(a) $ n = 1.5 $, 
<br>
(b) $ n = 1.0 $, and
<br>
(c) $ n = 0 $. 

## Solution

- **Known:** A gas in a piston–cylinder assembly undergoes an expansion for which $ P \cdot V^n = \mathrm{constant} $.
<br>
- **Find:** Evaluate the work if (a) $ n = 1.5 $, (b) $ n = 1.0 $, (c) $ n = 0 $.
<br>
- **Schematic and Given Data:** The given *P-V* relationship and the given data for pressure and volume can be used to construct the *pressure–volume* diagram of the process.
<br>
- **Engineering Model:**
> 1. The gas is a closed system.
> 2. The moving boundary is the only work mode.
> 3. The expansion is a polytropic process.
- **Analysis:**

**Initializing Python:**
<br>
In order to obtain the graphs using Python, **NumPy**, a package for scientific computing and **Matplotlib**, a library for visualization are imported into the notebook.

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Initial Values:**
<br>
The given data are the initial pressure $(P_1)$ and volume $(V_1)$, as well as the final volume $(V_2)$.
<br>
First, is necesary to convert $P_1$  from $\mathrm{bar}$ to $\mathrm{kPa}$:
<br>
$$ P_1 = 3 \, \mathrm{bar} \cdot \frac{100 \, \mathrm{kPa}}{1 \, \mathrm{bar}} = 300 \, \mathrm{kPa} $$

```python
P1 = 300 # kPa
print('P1:', P1,'kPa')
```

```python
V1 = 0.1 # m³
print('V1:', V1,'m³')
```

```python
V2 = 0.2 # m³
print('V2:', V2,'m³')
```

To make the graphs in Python, the best option is to create an array for the volume going from $V_1$ to $V_2$. One way to do it is using the function *linspace* of *NumPy*. This function needs the initial and final values, as well as the number of points that will compound the array. 

```python
points = 100
```

```python
V = np.linspace(V1, V2, points)
print('Volume:', V, 'm³')
```

### Polytropic process, *n* = 1.5

First, the value of *n* is initialized:

```python
n_poly = 1.5
print('n:', n_poly)
```

**Constant:**
<br>
The constant $C$ can be obtained using the values $P_1$ and $V_1$  using the given *pressure–volume* relation:

$$ C = P_1 \cdot V^n_1 $$

```python
C_poly = P1 * np.power(V1,n_poly)
print('C:', np.round(C_poly,3))
```

**Pressure @ state 2:**
<br>
To evaluate the work along this process, the pressure at state 2 is required. This can be found by using the *pressure–volume* relation, which on rearrangement yields:

$$ P_2 \cdot V^n_2 = C $$

$$ P_2 = \frac{C}{V^n_2} $$

```python
P2_poly = C_poly / np.power(V2,n_poly)
print('P2:', np.round(P2_poly,1),'kPa')
```

**Work:**
<br>
The expression for the work in a polytropic process is:

$$ W_{polytropic} = \frac{P_2 V_2 - P_1 V_1}{1-n} $$

This expression is valid for all values of *n* except $ n = 1.0 $.


```python
W_poly = (P2_poly*V2 - P1*V1)/(1 - n_poly)
print('W:', np.round(W_poly,1),'kJ')
```

***P-V* Diagram:**
<br>
To create the *P-V* diagram for this process, is necessary to create a new array for the values of pressure ranging from $P_1$ to $P_2$

$$ P = \frac{C}{V^n} \; \text{, for} \; P_1 < P < P_2 $$



```python
P_poly = C_poly / np.power(V,n_poly)
print('Pressure:', np.round(P_poly,2), 'kPa')
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}

plt.figure(figsize=(20,10))
plt.title('P-V Diagram · Polytropic Process, n = 1.5')
plt.rc('font', **font)
 
plt.plot(V,P_poly,'b--',linewidth=1.5)

plt.ylabel('Pressure (kPa)')
plt.xlabel('Volume (m³)')
plt.show()
```

### Isothermal process, *n* = 1

For $ n = 1.0 $, the *pressure–volume* relationship is:
$$ P \cdot V = C $$

```python
n_isot = 1
print('n:', n_isot)
```

**Constant:**
<br>
To evaluate the constant $C$, the values $P_1$ and $V_1$  are used in the same way as they were used in the polytropic process:

$$ C = P_1 \cdot V_1 $$

```python
C_isot = P1 * V1
print('C:', C_isot, 'kJ')
```

**Pressure @ state 2:**
<br>
The pressure at state 2 can be found by using the *pressure–volume* relation, which on rearrangement yields:

$$ P_2 = \frac{C}{V_2} $$

```python
P2_isot = C_isot / V2
print('P2:', np.round(P2_isot,1), 'kPa')
```

**Work:**
<br>
The work in an isothermal process is:

$$ W_{isothermal}= C \cdot \ln \left( \frac{V_2}{V_1} \right) $$


```python
W_isot = C_isot * np.log(V2/V1)
print('W:', np.round(W_isot,1),'kJ')
```

***P-V* Diagram:**
<br>
The pressure varies with the volume following the relationship:
$$ P = \frac{C}{V} $$

```python
P_isot = C_isot / V
print('Pressure:', np.round(P_isot,2), 'kPa')
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}

plt.figure(figsize=(20,10))
plt.title('P-V Diagram · Isothermal Process, n = 1')
plt.rc('font', **font)

plt.plot(V,P_isot,'r--',linewidth=1.5)

plt.ylabel('Pressure (kPa)')
plt.xlabel('Volume (m³)')
plt.show()
```

### Isobaric process, *n* = 0

For $ n = 0 $, the *pressure–volume* relation reduces to :
$$ P = \mathrm{constant} $$

**Pressure @ state 2:**
<br>
The pressure at state 2 will be the same as the pressure at state 1:

$$ P_2 = P_1 $$

```python
P2_isob = P1
print('P2:', P2_isob,'kPa')
```

**Work:**
<br>
With constant pressure, work becomes:

$$ W_{isobaric}= P_1 \cdot \left( V_2 - V_1 \right) $$


```python
W_isob = P1 * (V2 - V1)
print('W:', np.round(W_isob,1),'kJ')
```

***P-V* Diagram:**
<br>
The relationship between pressure and volume can be written as:
$$ P = P_1 \cdot V^0 $$

```python
P_isob = P1 * np.power(V,0)
print('Pressure:', P_isob, 'kPa')
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}

plt.figure(figsize=(20,10))
plt.title('P-V Diagram · Isobaric Process, n = 0')
plt.rc('font', **font)

plt.plot(V,P_isob,'g--',linewidth=1.5)

plt.ylabel('Pressure (kPa)')
plt.xlabel('Volume (m³)')
plt.show()
```

### Isochoric process, *n* = $\infty$

For $n = \infty$, the *pressure–volume* relation reduces to

$$ V = \mathrm{constant} $$

In this case, the array is created for the pressure.

```python
P_isoc = np.linspace(P2_poly, P2_isob, points)
print('Pressure:', np.round(P_isoc,2), 'kPa')
```

**Volume:**
<br>
Volume through the process is the same as that of state 2:

$$ V = V_2 $$

```python
print('V:', np.round(V2,1),'m³')
```

**Work:**
<br>
In an isochoric process, where volume is constant, work is zero:

$$ W_{isochoric} = 0 $$

***P-V* Diagram:**
<br>
The relationship between pressure and volume is, in this case:

$$ V = V_2 \cdot P^0 $$

```python
V_isoc =  V2 * np.power(P_isoc, 0)
print(V_isoc)
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}

plt.figure(figsize=(20,10))
plt.title('P-V Diagram · Isochoric Process, n = ∞')
plt.rc('font', **font)

plt.plot(V_isoc,P_isoc,'g--',linewidth=1.5)

plt.ylabel('Pressure (kPa)')
plt.xlabel('Volume (m³)')
plt.show()
```

### *P-V* Diagram for all the processes

In each case, the work for the process can be interpreted as the area under the curve representing the process on the accompanying *P–V* diagram. The relative areas are in agreement with the numerical results.

 Process   | $n$  | $P_2$ (kPa)  | $W$ (kJ) |
-----------|------|--------------|-----------
Polytropic |  1,5 |    106,1     |   17,6
Isothermal |  1,0 |    150       |   20,8
Isobaric   |  0   |    300       |   30
Isochoric  |  ∞   |              |    0

```python
font = {'family' : 'Helvetica',
        'size'   : 22}

plt.figure(figsize=(20,10))
plt.title('P-V Diagram')
plt.rc('font', **font)

plt.axis([0.095, 0.205, 0, 305])

plt.plot(V,P_isob,'g--',linewidth=1.5,label='isobaric')
plt.plot(V,P_isot,'r--',linewidth=1.5,label='isothermal')
plt.plot(V,P_poly,'b--',linewidth=1.5,label='polytropic')
plt.plot(V_isoc,P_isoc,'y--',linewidth=1.5,label='isochoric')

plt.ylabel('Pressure (kPa)')
plt.xlabel('Volume (m³)')
plt.legend()
plt.show()
```

In each of the cases considered, it is not necessary to identify the gas (or liquid) contained within the piston–cylinder assembly. The calculated values for *W* are determined by the process path and the end states. However, if it is desired to evaluate a property such as temperature, both the nature and amount of the substance must be provided because appropriate relations among the properties of the particular substance would then be required.
