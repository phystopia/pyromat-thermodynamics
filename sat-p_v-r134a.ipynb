```python
import pyromat as pm
import numpy as np
import matplotlib.pyplot as plt

pm.config["unit_pressure"] = "kPa"
pm.config["def_p"] = 100

R134a = pm.get("mp.C2H2F4")

p = np.arange(0.4,5000,1000)
```

```python
T = R134a.Ts(p=p)
s = R134a.ss(p=p)
d = R134a.ds(p=p)
vf= 1/d[0]
vg= 1/d[1]
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(15,10))
plt.xscale("log")
plt.yscale("log")
plt.title('Saturation P-v Diagram for Tetrafluoroethane (R-134a)')
plt.rc('font', **font)
plt.plot(vf,p, 'b--')
plt.plot(vg,p,'r--')
plt.ylabel('Pressure (kPa)')
plt.xlabel('specific volume (v)')
```
