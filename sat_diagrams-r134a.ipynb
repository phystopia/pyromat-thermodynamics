```python
import pyromat as pm
import numpy as np
import matplotlib.pyplot as plt

pm.config["unit_pressure"] = "kPa"

R134a = pm.get("mp.C2H2F4")

p = np.arange(0.4,4059)
```

## Saturation P-v Diagram for Tetrafluoroethane (R-134a)

```python
d = R134a.ds(p=p)
vf= 1/d[0]
vg= 1/d[1]
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(15,10))
plt.xscale("log")
plt.yscale("log")
plt.title('Saturation P-v Diagram for Tetrafluoroethane (R-134a)')
plt.rc('font', **font)
plt.plot(vf,p, 'b--')
plt.plot(vg,p,'r--')
plt.ylabel('Pressure (kPa)')
plt.xlabel('specific volume (v)')
```

```python
## Saturation T-s Diagram for Tetrafluoroethane (R-134a)
```

```python
T = R134a.Ts(p=p)
s = R134a.ss(p=p)
```

```python
font = {'family' : 'Helvetica',
        'size'   : 22}
plt.figure(figsize=(15,10))
# plt.xscale("log")
# plt.yscale("log")
plt.title('Saturation T-s Diagram for Tetrafluoroethane (R-134a)')
plt.rc('font', **font)
plt.plot(s[0], T, 'b--')
plt.plot(s[1], T,'r--')
plt.ylabel('Temperature (kPa)')
plt.xlabel('specific enthropy (kJ/kg)')
```
